```python
#luas => p*l
#keliling => 2 * (p+l)

p = input('masukan angka pada pesegi')
l   = input('masukan angka pada persegi')

Luas    = int(p) * int(l)
Keliling = 2 * (int(p) + int(l))

print('Luas Persegi Panjang',Luas)
print('keliling dari pesegi panjang',Keliling)
```

```python
#isikan nama anda : Nadzar Hadi
# isikan tanggal lahir : 14/06/2002
#OUTPUT
#nama awal : nadzar
#nama akhir : hadi
#tgl lahir : 1
#bulan lahir : 06
#tahun lahir : 2002
#nama inisial : NH

nama = input("Masukan nama anda")
tanggal = input("Masukan tanggal lahir")

list_nama = nama.split()
list_tanggal = tanggal.split("/")
print("nama awal :" ,list_nama[0])
print("nama akhir :" ,list_nama[1])
print("tanggal lahir :" ,list_tanggal[0])
print("bulan lahi :" ,list_tanggal[1])
print("tahun lahir :" ,list_tanggal[2])


inisial = list_nama[0][0] + list_nama[1][0]
print("nama inisial :", inisial)
```
