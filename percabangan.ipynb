```python
nilai = int(input('isikan bilangan :'))
sisa_bagi = nilai % 2

if sisa_bagi ==0:
    print(nilai,"adalah bilangan genap")
else:
    print(f"{nilai} adalah nilangan ganjil")
```

```python
#0 - 49 => E
#50 - 59 => D
#60 - 69 => C
#70 - 84 => B
#85 - 100 => A

nilai_mhs = int(input('isikan nilai pemograman mahasiswa:'))
keterangan =""
if nilai_mhs >=0 and nilai_mhs < 50:
    keterangan = "E"
elif nilai_mhs >=50 and nilai_mhs <60:
    keterangan = "D"
elif nilai_mhs >=60 and nilai_mhs <70:
    keterangan = "C"
elif nilai_mhs >=70 and nilai_mhs <85:
    keterangan = "B"
elif nilai_mhs >=85 and nilai_mhs <101:
    keterangan = "A"
print("nilai pemograman mahasiswa :",keterangan)
```
